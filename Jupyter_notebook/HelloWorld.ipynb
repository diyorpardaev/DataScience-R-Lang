# **Rda asosiy ishlatiladigan arfimetik belgilarning qisqacha mazmumi**


*  (+)  qo'shish  3+5+8
*  (-)  ayirish   9-3
*  (*)  ko'paytirish 7*5
*  (/)  bo'lish   8/3
*  (%%) qoldiqni topish  8%%3
*  (^)  daraja  2^3

Oxirgi qatordagi 2^3  bu 2 ning uchinchi darajasini bildiradi ya'ni 2³.Buning qiymati 2*2*2 bilan teng.

Keyin Rda  huddi Pythondagidek # izoh belgisi bor.Yozilgan Kodga sharhlarni yozishda # panjara belgisi ishlatilib,undan keyin izoh yozilsa umumiy kodga tasir qilmaydi.Izohlar muallifga keyinchalik R dasturini qqayta ko'rib chiqqanda yoki uni uchinchi shaxs ko'rganda dastur mazmunini osongina tushunishga yordam beradi.

# Arfimetik amallar funksiyalarda
R da asosan biz  raqamlar va arfimetik operatorlar yordamida amallarni bajarishimiz mumkin.Lekin hisob-kitoblarni qulay bajarishimiz uchun maxsus funksiyalardan foydalanishimiz kerak.Keling quyidagi kodda bir nechtasini ko'rib chiqamiz:

```r
log(10)+5        #logarifmik funksiya
sqrt(25)         #kvadrat ildiz
max(5,3,2)       #eng katta qiymat
```

Yuqoridagi kodda mos ravishda logarifmik funksiya ,kvadrat ildiz ca eng katta qiymatni topish funksiyasi qo'llanilgan.

Quyida esa Rda tez-tez ishlatiladigan arfimetik funksiyalarning qisqacha mazmuni bilan tanishamiz

* log() - logarifmik funksiya - log(10), logarifm 2 asosga ko'ra 10
* sqrt() - Kvadrat ildiz - sqrt(36)
* max() - Eng katta qiymatni topish - max(3,9,5)
* min() - eng kichik qiymatni topish - min(3,9,5)
* abs() - abstrakt funksiya - abs(-10)
* factorial - faktorial - factorial(5)
* sin(),cos(),tan() - Trigonometrik funksiyalar - sin(pi/2)
